## Data Structures in Python

Data structure is a specific way of storing and organizing data in memory such that it can be retrieved and used in a most productive way. 

The data structure plays an important role in code optimization. The rate at which the data is stored or retrieved, from the memory, is quite important in algorithm optimization especially when huge amount of data are involved.

#### The following are the data structures in python that are mainly used for data science related projects.



### 1) Lists


A list is a collection of homogeneous/heterogenous elements (int, float, string, etc.)

```python
# Create an empty list

Empty_List = []
```

```python
print(Empty_List)
```

```python
len(Empty_List)
```

```python
# create a simple list

list_var = ['hello', 'how', 'are', 'you', 1, 10, 'how', 'well', 'are'] 
```

```python
# Display the content of the List ( whole list )

print(list_var)
```

```python
len(list_var)
```

#### Functions available for List data structure 

```python
# Find the data structure type of a list variable 

type(list_var)
```

```python
# Find the count of elements in a list 

len(list_var)
```

##### Access the specific elements with in a List 


Note - The index in Python starts from 0 ( and not 1 )



```python
list_var
```

```python
# Accessing the first element in the list 

list_var[0]
```

```python
# Accessing the 4th element in the list 

list_var[3]
```

```python
# Accessing the last element in the list 

list_var[-1]
```

```python
# Accessing the 3rd last element in the list 

list_var[-3]
```

```python
# Accessing the elements from 1st position till 5th position [approach - 1]

list_var[0:5]
```

```python
# Accessing the elements from 1st position till 5th position [approach - 2]

list_var[:5]
```

```python
# Accessing the elements from 3rd position till 5th position 

list_var[2:5]
```

```python
# Accessing the elements from 4th position till end of the list

list_var[3:]
```

### Some advanced element accessing from a list


Slicing

    To get a specific set of elements from the list

```python
list_var
```

```python
list_var[0:8:2]
```

```python
# Accessing the elements which are in odd positions [approach 1]

list_var[0::2]
```

```python
# Accessing the elements which are in odd positions [approach 2]

list_var[::2]
```

```python
list_var
```

```python
# Accessing the elements which are in even positions 

list_var[1::2]
```

```python
# Accessing the elements in reverse order [approach 1]

list_var[-1::-1]
```

```python
# Accessing the elements in reverse order [approach 2]

list_var[::-1]
```
